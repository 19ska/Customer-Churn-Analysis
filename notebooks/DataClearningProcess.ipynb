```python
import pandas as pd
```

```python
file = pd.read_csv("WA_Fn-UseC_-Telco-Customer-Churn.csv")
```

```python
#One hot encoding of column(gender)
#Male = 1
#Female = 0

file['gender'] = file['gender'].replace({'Female':0, 'Male':1})

file.head()

file
```

```python
#One hot encoding of column(PhoneService)
# Yes = 1
# No = 0

file['PhoneService'] = file['PhoneService'].replace({'No':0,'Yes':1})

file.head()
```

```python
#One hot encoding of column (Patners)

#Yes = 1
#No = 0

file['Partner'] = file["Partner"].replace({'No':0,'Yes':1})

file.head()
```

```python
#one hot encoding of column (Dependents)

file['Dependents'] = file['Dependents'].replace({'No':0,'Yes':1})

file.head()
```

```python
#One hot encoding of column Churn

file['Churn'] = file['Churn'].replace({'No':0, 'Yes':1})
file.head()
```

```python
#One hot encoding of column(PaperlessBilling)
# Yes = 1
# No = 0

file['PaperlessBilling'] = file['PaperlessBilling'].replace({'No':0,'Yes':1})

file.head()
```

```python
#Categorical Feature Engineering on Column MultipleLines

one_hot_encoded_MultipleLines = pd.get_dummies(file,columns=["MultipleLines"], prefix="MultipleLines", prefix_sep="_")
one_hot_encoded_MultipleLines.head()
```

```python
#Adding the new columns to the original dataframe

file = one_hot_encoded_MultipleLines
file.head()
```

```python
#Checking the columns in the dataframe

file.columns
```

```python
#Categorical Feature Engineering on Column InternetService

file = pd.get_dummies(file,columns=["InternetService"], prefix="InternetService", prefix_sep="_")
file.head()
```

```python
#Categorical Feature Engineering on Column OnlineSecurity

file = pd.get_dummies(file,columns=["OnlineSecurity"], prefix="OnlineSecurity", prefix_sep="_")
file.head()
file.columns
```

```python
file.head()
```

```python
#Categorical Feature Engineering on Column Online backup

file = pd.get_dummies(file,columns=["OnlineBackup"], prefix="OnlineBackup", prefix_sep="_")
file.head()
```

```python
#Categorical Feature Engineering on Column TechSupport

file = pd.get_dummies(file,columns=['TechSupport'],prefix='TechSupport',prefix_sep='_')
file.head()
```

```python
#Categorical Feature Engineering on Column StreamingTV

file = pd.get_dummies(file,columns=['StreamingTV'],prefix='StreamingTV',prefix_sep='_')
file.head()
```

```python
#Categorical Feature Engineering on Column StreamingMovies

file = pd.get_dummies(file,columns=['StreamingMovies'],prefix="StreamingMovies",prefix_sep="_")
file.head()
```

```python
#Categorical Feature Engineering on Column Contract

file = pd.get_dummies(file,columns=['Contract'],prefix='Contract',prefix_sep="_")
file.head()
```
